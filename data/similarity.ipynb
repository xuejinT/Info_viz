```python
import pandas as pd
import numpy as np
from collections import OrderedDict
import time
import pickle
import multiprocessing
import json
```

```python
df = pd.read_csv("US_final.csv",sep="\\")
```

```python
unique_cate = df.category_name.unique()
for cate in unique_cate:
    curr_df = df[df['category_name']==cate]
    print(cate,len(curr_df))
```

```python
#1. add a column of tags
#2. search in same channel
#3. search in all 
#4. merge
```

```python
df_withKeywords = df.copy()
df_withKeywords["keywords"] = ""
```

```python

for i, row in df_withKeywords.iterrows():
    keywords = set()
    cleaned_keywords = set()
    keywords.update(row.title.split(" "))
    keywords.update(row.tags.split("|"))
    keywords.update(str(row.description).split(" "))
    
    for word in keywords:
        word = word.lower()
        for char in word:
            if not char.isalpha() and not char.isnumeric():
                word = word.replace(char,'')
        if len(word)>1:
            cleaned_keywords.add(word)
    
    df_withKeywords.at[i,"keywords"] = "|".join(list(cleaned_keywords))
    
```

```python
# sort_map =[]

def get_sim_score(row1, row2):
    keywords1 = row1.keywords.split("|")
    keywords2 = row2.keywords.split("|")
    sim_count = 0
    for word1 in keywords1:
        for word2 in keywords2:
            if word1 == word2:
                sim_count+=1
    return sim_count

def search_sim_channel(curr_row, curr_df):
    sim_videos = {}
    for i, row in curr_df.iterrows():
        if(row.video_id != curr_row.video_id):
            sim_score = get_sim_score(row, curr_row)
            sim_videos[row.video_id] = sim_score
    sort_map = OrderedDict(sorted(sim_videos.items(), key=lambda item: item[1],reverse=True))
    return sort_map
```

```python
count =0
sim_map={}

error = []
unique_id = df.video_id.unique()
```

```python
df1 = df_withKeywords.iloc[:len(df_withKeywords)//2, :]
df2 = df_withKeywords.iloc[len(df_withKeywords)//2:, :]

for vid in unique_id:
    sim_map[vid] = {}
    
for i, curr_row in df1.iterrows():
    curr_id = curr_row.video_id
    if count%10 == 0:
        time1 = time.time()
    try:
        sim_videos = search_sim_channel(curr_row,df_withKeywords)
        sim_map[curr_id].update(sim_videos)
    except Exception as e:
        error+=[curr_id]
    count+=1
    if count%10 == 0:
        avg_time = (time.time()-time1)/10
        est_rest = round((len(df1)//2-count)*avg_time/3600,2)
        print(count,"/",len(df1), "--",est_rest,"hrs left",'--',error)
```
